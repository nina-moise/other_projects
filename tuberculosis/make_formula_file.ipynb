**Описание:** Этот notebook создает в указанной папке **path** файл **indicators.jsonl** (формат jsonl) с формулами для расчета показателей заболеваемости туберкулезом. 

**Формат выходного файла:**

Одна строка — один объект, объекты в JSONL не разделяются запятыми, разделитель — переход строки.

- id - словарь - идентификатор
- name - словарь - полное наименование показателя
- short_name - словарь - сокращенное название показателя
- numerator - числитель - список из одного или нескольких источников. Это позволяет просуммировать данные из разных таблиц.
- denominator - знаменатель - список из одного или нескольких источников (для суммы данных из разных таблиц) или тип населения из демографических данных.
- multiplier - множитель

numerator и denominator - это списки со следующей структурой:
- year - год, может принимать значение "current" (текущий) или "last"(прошлый)
- table -  номер таблицы в отчетной форме
- row - строка в таблице
- col - колонка в таблице

row и col - однозначно определяют ячейку таблицы отчетной формы.

Кроме того, если в знаменателе для расчета требуется количество соответствующего населения, то структура словаря для denominator может быть следующая:
- "population": название столбца из связанной таблицы population, содержащий количество соответствующего для расчета показателя населения за текущий год.
  

**Главное преимущество** такой структуры — её масштабируемость. Т.к. numerator и denominator являются списками, система сможет обрабатывать как простые показатели, так и сложные агрегаты без изменения кода парсера. Можно использовать эту структуру для расчета других аналогичных показателей по статистическим медицинским формам.

```python
# Создаем пустой файл (или затираем старый, если он был)
with open(file_name, "w", encoding="utf-8") as f:
    pass 
```

```python
def add_indicator(data_object):
    """ Добавляем строку с формулой расчета в конец json-файла-справочника
    """
    with open(file_name, "a", encoding="utf-8") as f:
        # Сериализуем словарь в строку и добавляем перенос строки \n
        json_record = json.dumps(data_object, ensure_ascii=False)
        f.write(json_record + "\n")
```
